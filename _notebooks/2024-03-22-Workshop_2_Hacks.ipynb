## Question 1 - Pojos and Access Control:
Situation: The school librarian wants to create a program that stores all of the books within the library in a database and is used to manage their inventory of books available to the students. You decided to put your amazing code skills to work and help out the librarian!

a. Describe the key differences between the private and public access controllers and how it affects a POJO
**Private access modifiers in POJOs restrict access to fields or methods within the same class, ensuring data hiding and preventing external interference, as they can only be accessed within the class they are declared in. On the other hand, public access modifiers make fields or methods accessible from any other class, which can lead to a loss of control over data integrity and increases the risk of unintended modifications, as they can be freely accessed and modified by any part of the program.**

b. Identify a scenario when you would use the private vs. public access controllers that isn't the one given in the scenario above
**In a banking application, private access controllers would secure sensitive data like account balances, while public access controllers would enable methods like deposit and withdraw to be accessible for account interactions.**

Create a Book class that represents the following attributes about a book: title, author, date published, person holding the book and make sure that the objects are using a POJO, the proper getters and setters and are secure from any other modifications that the program makes later to the objects
**See Below**


```java
public class Book {
    private String title;
    private String author;
    private String datePublished;
    private String personHolding;

    // Constructors
    public Book() {
    }

    public Book(String title, String author, String datePublished, String personHolding) {
        this.title = title;
        this.author = author;
        this.datePublished = datePublished;
        this.personHolding = personHolding;
    }

    // Getters and Setters
    public String getTitle() {
        return title;
    }

    public void setTitle(String title) {
        this.title = title;
    }

    public String getAuthor() {
        return author;
    }

    public void setAuthor(String author) {
        this.author = author;
    }

    public String getDatePublished() {
        return datePublished;
    }

    public void setDatePublished(String datePublished) {
        this.datePublished = datePublished;
    }

    public String getPersonHolding() {
        return personHolding;
    }

    public void setPersonHolding(String personHolding) {
        this.personHolding = personHolding;
    }
}

Book.main(null)
```

## Question 2 - Writing Classes:

(a) Describe the different features needed to create a class and what their purpose is.
**To create a class in Java, you need to define its features, including attributes (data variables), constructors (initialization methods), methods (actions or functionality), access modifiers (visibility), encapsulation (data hiding), inheritance (property and behavior inheritance), polymorphism (single action in multiple ways), and documentation (comments explaining purpose), all contributing to a maintainable codebase.***

(b) Code:

Create a Java class BankAccount to represent a simple bank account. This class should have the following attributes:
- accountHolder (String): The name of the account holder.
balance (double): The current balance in the account.
Implement the following mutator (setter) methods for the BankAccount class:
- setAccountHolder(String name): Sets the name of the account holder.
- deposit(double amount): Deposits a given amount into the account.
- withdraw(double amount): Withdraws a given amount from the account, but only if the withdrawal amount is less than or equal to the current balance.
Ensure that the balance is never negative.

```java
// Constructor
public BankAccount(String accountHolder, double balance) {
    this.accountHolder = accountHolder;
    this.balance = balance;
}

// Mutator methods
public void setAccountHolder(String name) {
    this.accountHolder = name;
}

public void deposit(double amount) {
    if (amount > 0) {
        balance += amount;
    }
}

public void withdraw(double amount) {
    if (amount > 0 && amount <= balance) {
        balance -= amount;
    }
}

// Getter method for balance
public double getBalance() {
    return balance;
}

BankAccount.main(null)
```

## Question 3 - Instantiation of a Class

(a) Explain how a constructor works, including when it runs and what generally is done within a constructor.

**A constructor in object-oriented programming is a special type of method that is automatically called when an instance of a class is created. Its main purpose is to initialize the newly created object. Constructors have the same name as the class and typically do not have a return type specified.**

(b) Create an example of an overloaded constructor within a class. You must use at least three variables. Include the correct initialization of variables and correct headers for the constructor. Then, run the constructor at least twice with different variables and demonstrate that these two objects called different constructors. 

```java
public class Example {
    private int var1;
    private double var2;
    private String var3;
    
    // Default constructor
    public Example() {
        this.var1 = 0;
        this.var2 = 0.0;
        this.var3 = "";
        System.out.println("Default constructor called.");
    }
    
    // Overloaded constructor
    public Example(int var1, double var2, String var3) {
        this.var1 = var1;
        this.var2 = var2;
        this.var3 = var3;
        System.out.println("Overloaded constructor called.");
    }
    
    public void display() {
        System.out.println("var1: " + var1 + ", var2: " + var2 + ", var3: " + var3);
    }
    
    public static void main(String[] args) {
        // Creating objects using different constructors
        Example obj1 = new Example(); // Default constructor called
        obj1.display(); // Displaying values
        
        Example obj2 = new Example(10, 3.14, "Hello"); // Overloaded constructor called
        obj2.display(); // Displaying values
    }
}
Example.main(null)
```

## Question 4 - Wrapper Classes:

(a) Provide a brief summary of what a wrapper class is and provide a small code block showing a basic example of a wrapper class.

**A wrapper class in Java is a class that wraps around a primitive data type and provides additional functionality. It allows primitive data types to be used as objects. Wrapper classes are useful when you need to treat primitive types as objects, such as when working with collections or using certain methods that require objects. For each primitive data type in Java, there is a corresponding wrapper class.**

(b) Create a Java wrapper class called Temperature to represent temperatures in Celsius. Your Temperature class should have the following features:

Fields:

A private double field to store the temperature value in Celsius.


Constructor:

A constructor that takes a double value representing the temperature in Celsius and initializes the field.


Methods:

getTemperature(): A method that returns the temperature value in Celsius.
setTemperature(double value): A method that sets a new temperature value in Celsius.
toFahrenheit(): A method that converts the temperature from Celsius to Fahrenheit and returns the result as a double value. 

```java
public class Temperature {
    private double celsius;

    // Constructor
    public Temperature(double celsius) {
        this.celsius = celsius;
    }

    // Getter method
    public double getTemperature() {
        return celsius;
    }

    // Setter method
    public void setTemperature(double celsius) {
        this.celsius = celsius;
    }

    // Convert Celsius to Fahrenheit
    public double toFahrenheit() {
        return (celsius * 9 / 5) + 32;
    }

    public static void main(String[] args) {
        // Create an instance of Temperature
        Temperature temp = new Temperature(25.0);

        // Display the temperature in Celsius
        System.out.println("Temperature in Celsius: " + temp.getTemperature());

        // Convert and display the temperature in Fahrenheit
        System.out.println("Temperature in Fahrenheit: " + temp.toFahrenheit());

        // Change the temperature and display again
        temp.setTemperature(30.0);
        System.out.println("New temperature in Celsius: " + temp.getTemperature());
        System.out.println("New temperature in Fahrenheit: " + temp.toFahrenheit());
    }
}
Temperature.main(null)
```

## Question 5 - Inheritence:

Situation: You are developing a program to manage a zoo, where various types of animals are kept in different enclosures. To streamline your code, you decide to use inheritance to model the relationships between different types of animals and their behaviors.

(a) Explain the concept of inheritance in Java. Provide an example scenario where inheritance is useful.

**Inheritance in Java is a mechanism where a new class, called a subclass or derived class, is created using the properties and behaviors of an existing class, called a superclass or base class. The subclass inherits all the fields and methods of the superclass and can also have its own additional fields and methods. This allows for code reuse, abstraction, and the creation of a hierarchical structure among classes. An example scenario where inheritance is useful is in modeling different types of vehicles. You can have a superclass called Vehicle with common attributes and methods shared by all vehicles, such as speed, color, and a method to start the vehicle. Then, you can have subclasses like Car, Bicycle, and Truck, each with specific attributes and methods unique to that type of vehicle. This way, you avoid redundancy by defining common functionalities once in the superclass and specialize them as needed in subclasses.**

(b) Code:

You need to implement a Java class hierarchy to represent different types of animals in the zoo. Create a superclass Animal with basic attributes and methods common to all animals, and at least three subclasses representing specific types of animals with additional attributes and methods. Include comments to explain your code, specifically how inheritance is used.


```java
// Superclass representing generic animal
class Animal {
    private String species;
    private int age;
    
    // Constructor
    public Animal(String species, int age) {
        this.species = species;
        this.age = age;
    }
    
    // Getter and setter methods
    public String getSpecies() {
        return species;
    }
    
    public void setSpecies(String species) {
        this.species = species;
    }
    
    public int getAge() {
        return age;
    }
    
    public void setAge(int age) {
        this.age = age;
    }
    
    // Method to make sound
    public void makeSound() {
        System.out.println("The animal makes a sound.");
    }
}

// Subclass representing mammals
class Mammal extends Animal {
    private String furColor;
    
    // Constructor
    public Mammal(String species, int age, String furColor) {
        super(species, age);
        this.furColor = furColor;
    }
    
    // Getter and setter methods for fur color
    public String getFurColor() {
        return furColor;
    }
    
    public void setFurColor(String furColor) {
        this.furColor = furColor;
    }
    
    // Method to display mammal information
    public void displayInfo() {
        System.out.println("Species: " + getSpecies());
        System.out.println("Age: " + getAge());
        System.out.println("Fur Color: " + furColor);
    }
}

// Subclass representing birds
class Bird extends Animal {
    private String featherColor;
    
    // Constructor
    public Bird(String species, int age, String featherColor) {
        super(species, age);
        this.featherColor = featherColor;
    }
    
    // Getter and setter methods for feather color
    public String getFeatherColor() {
        return featherColor;
    }
    
    public void setFeatherColor(String featherColor) {
        this.featherColor = featherColor;
    }
    
    // Method to display bird information
    public void displayInfo() {
        System.out.println("Species: " + getSpecies());
        System.out.println("Age: " + getAge());
        System.out.println("Feather Color: " + featherColor);
    }
}

// Subclass representing reptiles
class Reptile extends Animal {
    private boolean hasScales;
    
    // Constructor
    public Reptile(String species, int age, boolean hasScales) {
        super(species, age);
        this.hasScales = hasScales;
    }
    
    // Getter and setter methods for scales
    public boolean hasScales() {
        return hasScales;
    }
    
    public void setHasScales(boolean hasScales) {
        this.hasScales = hasScales;
    }
    
    // Method to display reptile information
    public void displayInfo() {
        System.out.println("Species: " + getSpecies());
        System.out.println("Age: " + getAge());
        System.out.println("Has Scales: " + hasScales);
    }
}

// Main class to demonstrate inheritance
public class Zoo {
    public static void main(String[] args) {
        // Create instances of different animals
        Mammal lion = new Mammal("Lion", 5, "Yellow");
        Bird parrot = new Bird("Parrot", 3, "Green");
        Reptile snake = new Reptile("Snake", 2, true);
        
        // Display information about the animals
        System.out.println("Information about lion:");
        lion.displayInfo();
        lion.makeSound();
        
        System.out.println("\nInformation about parrot:");
        parrot.displayInfo();
        parrot.makeSound();
        
        System.out.println("\nInformation about snake:");
        snake.displayInfo();
        snake.makeSound();
    }
}

Zoo.main(null)
```
